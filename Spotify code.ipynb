### Top 10 Artist who has highest listen time 

```python
data["duration_min"]=data.duration_ms/60000
top_listened_artists=data.groupby("artist").duration_min.sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(18,5))
plt.bar(x=top_listened_artists.index,height=top_listened_artists.values)
for x,y in zip (top_listened_artists.index,top_listened_artists.values):
    plt.text(x,y+.8,round(y,2),horizontalalignment="center")
plt.title("Top 10 Artist who has highest listen time",size=20)
plt.show()
```

### Top 10 songs who has highest listen time

```python
top_song = data.sort_values(by="duration_min",ascending=False).head(10)
plt.figure(figsize=(20,7.5))
plt.barh(y=top_song.song_title,width=top_song.duration_min,color='r')
for x,y in zip(top_song.song_title,top_song.duration_min):
    plt.text(y,x,round(y,2)) 
plt.title("Top 10 songs who has highest listen time",size=25)    
plt.show()    
```

### Most danceable song

```python
danceable_song=data.sort_values(by="danceability",ascending=False).head(10)
plt.figure(figsize=(18,5))
plt.bar(x=danceable_song.song_title,height=danceable_song.danceability)
for x,y in zip (danceable_song.song_title,danceable_song.danceability):
    plt.text(x,y+0.01,y)
plt.title("Most danceable song",size=20)  
plt.show
```

### Most energetic songs

```python
energetic=data.sort_values(by="energy",ascending=False).head(10)
plt.figure(figsize=(19,5))
plt.bar(x=energetic.song_title,height=energetic.energy)
for x,y in zip (energetic.song_title,energetic.energy):
    plt.text(x,y+.01,y)
plt.title("Most energetic song",size=20)
plt.show() 
```

### Loudest song

```python
loudest=data.sort_values(by="loudness",ascending=False).head(10)
plt.figure(figsize=(21,5))
plt.bar(x=loudest.song_title,height=loudest.loudness)
for x,y in zip (loudest.song_title,loudest.loudness):
    plt.text(x,y-.06,y)
plt.title("Loudest song",size=20)
plt.show()
```

### 10 Fastest songs 

```python
fastest_song=data.sort_values(by="tempo",ascending=False).head(10)
plt.figure(figsize=(20,8))
plt.bar(x=fastest_song.song_title,height=fastest_song.tempo,color='g')
for x,y in zip (fastest_song.song_title,fastest_song.tempo):
    plt.text(x,y+3,y,horizontalalignment='center')
plt.title("10 Fastest songs",size=20)
plt.show()
```

### 10 slowest songs

```python
slowest=data.sort_values(by='tempo',ascending=True).head(10)
plt.figure(figsize=(20,8))
plt.barh(y=slowest.song_title,width=slowest.tempo)
for x,y in zip(slowest.song_title,slowest.tempo):
    plt.text(y,x,y)
plt.title("10 Slowest songs",size=20)
plt.show() 
```
